```python
import os
import torch
import matplotlib.pyplot as plt
```

```python
dynamics = 'linear'

plt.figure(figsize=(12,5))
METHODS = ['euler','rk4','fixed_adams','midpoint']
for method in METHODS:
    print(method)
    hist = torch.Tensor(torch.load(os.path.join(dynamics,method,'adjoint_'+method+'_loss.pth')))
    # print(torch.ones((10)).shape)
    # break
    smooth_hist = torch.nn.functional.conv1d(hist.unsqueeze(0), torch.ones((1,1,10))/10).squeeze(0)
    h = plt.semilogy(smooth_hist,alpha=1,label=method)
    plt.semilogy(hist,alpha=0.2,color=h[0].get_color())
    # plt.draw()
    # break
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('$1og_{10}~L_1$ loss')
plt.savefig(os.path.join(dynamics, '_'.join(METHODS)+'_adjoint_loss.pdf'))
```
