```python
import torchvision
import torch
```

```python

mnist = torchvision.datasets.MNIST(root='datasets',download=True,train=False)
```

```python
[torchvision.transforms.functional.pil_to_tensor(mnist[i][0]) for i in range(4)]

# make a function adds two numbers
```

```python

torchvision.utils.save_image([torchvision.transforms.functional.pil_to_tensor(mnist[i][0]).float() for i in range(32)],'out.pdf')
```
